```python
# Your implementation:

nutrition = pd.read_csv('breakfast-cereals-nutrition.csv')

product = pd.read_csv('breakfast-cereals-products.csv')

other = pd.read_csv('breakfast-cereals-other.csv')
```

### Visualization 5.3

**Actions:**
- Display a pie chart showing the percentage of cereals grouped by their shelf placement in supermarkets.
- Label the title, x-axis, and y-axis.

**Hints:**

- You can use Pandas, Matplotlib, or Seaborn to create this pie chart.

```python
# Your implementation:

figure, axis = plot.subplots()

axis.set_title=('Percentage of cerreals grouped by the shelf placement')

shelf = MasterDataFrame.groupby('SHELF')

count = shelf['SHELF'].size()

percent = (count / count.sum()) * 100

label = [1.0, 2.0, 3.0]

percent = list(percent)

axis.axis('equal')

axis.pie(count, labels = label, autopct='%.1f%%' )
```
